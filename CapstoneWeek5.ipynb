```python
def get_latlng(District):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Manila, Philippines'.format(District))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
# call the function to get the coordinates, store in a new list using list comprehension
coords = [ get_latlng(District) for District in df_MNL['District'].tolist() ]
coords
```

```python
df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])
```

```python
df_MNL['Latitude'] = df_coords['Latitude']
df_MNL['Longitude'] = df_coords['Longitude']
```

```python
radius = 2000
LIMIT = 100

venues = []

for lat, long, neighborhood in zip(df_MNL['Latitude'], df_MNL['Longitude'], df_MNL['District']):
    
    # create the API request URL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    # return only relevant information for each nearby venue
    for venue in results:
        venues.append((
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
# convert the venues list into a new DataFrame
df_venues = pd.DataFrame(venues)

# define the column names
df_venues.columns = ['District', 'Latitude', 'Longitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']

print(df_venues.shape)
df_venues.head()
```

```python
# one hot encoding
MNL_onehot = pd.get_dummies(df_venues[['VenueCategory']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
MNL_onehot['District'] = df_venues['District'] 

# move neighborhood column to the first column
fixed_columns = [MNL_onehot.columns[-1]] + list(MNL_onehot.columns[:-1])
MNL_onehot = MNL_onehot[fixed_columns]

print(MNL_onehot.shape)
MNL_onehot.head()
```

```python
MNL_grouped = MNL_onehot.groupby(["District"]).mean().reset_index()

print(MNL_grouped.shape)
MNL_grouped
```

```python
len(MNL_grouped[MNL_grouped["Grocery Store"] > 0])
```

```python
MNL_cs = MNL_grouped[["District","Grocery Store"]]
MNL_cs
```

```python
# set number of clusters
clusters = 3

MNL_clustering = MNL_cs.drop(["District"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=clusters, random_state=0).fit(MNL_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster as well as the top 10 venues for each district.
MNL_merged = MNL_cs.copy()

# add clustering labels
MNL_merged["Cluster Labels"] = kmeans.labels_
```

```python
MNL_merged.head()
```

```python
# merge toronto_grouped with toronto_data to add latitude/longitude for each district
MNL_merged = MNL_merged.join(df_MNL.set_index("District"), on="District")

print(MNL_merged.shape)
MNL_merged.head() # check the last columns!
```

```python
# sort the results by Cluster Labels
print(MNL_merged.shape)
MNL_merged.sort_values(["Cluster Labels"], inplace=True)
MNL_merged
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=13)

# set color scheme for the clusters
x = np.arange(clusters)
ys = [i+x+(i*x)**2 for i in range(clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(MNL_merged['Latitude'], MNL_merged['Longitude'], MNL_merged['District'], MNL_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
